```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
```

```python
df = pd.read_csv('./Virtual_Reality_in_Education_Impact.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
mean_age = df['Age'].mean()
print(f'Média da coluna Age: {mean_age}')
mode_age = df['Age'].mode()
print(f'Moda da coluna Age: {mode_age}')
std_age = df['Age'].std()
print(f'Desvio padrão da coluna Age: {std_age}')
```

```python
mean_hours = df['Hours_of_VR_Usage_Per_Week'].mean()
print(f'Média da coluna Hours_of_VR_Usage_Per_Week: {mean_hours}')
mode_hours = df['Hours_of_VR_Usage_Per_Week'].mode()
print(f'Moda da coluna Hours_of_VR_Usage_Per_Week: {mode_hours}')
std_hours = df['Hours_of_VR_Usage_Per_Week'].std()
print(f'Desvio padrão da coluna Hours_of_VR_Usage_Per_Week: {std_hours}')
```

```python
mean_engagement = df['Engagement_Level'].mean()
print(f'Média da coluna Engagement_Level: {mean_engagement}')
mode_engagement = df['Engagement_Level'].mode()
print(f'Moda da coluna Engagement_Level: {mode_engagement}')
std_engagement = df['Engagement_Level'].std()
print(f'Desvio padrão da coluna Engagement_Level: {std_engagement}')
```

```python
mean_perceived = df['Perceived_Effectiveness_of_VR'].mean()
print(f'Média da coluna Perceived_Effectiveness_of_VR: {mean_perceived}')
mode_perceived = df['Perceived_Effectiveness_of_VR'].mode()
print(f'Moda da coluna Perceived_Effectiveness_of_VR: {mode_perceived}')
std_perceived = df['Perceived_Effectiveness_of_VR'].std()
print(f'Desvio padrão da coluna Perceived_Effectiveness_of_VR: {std_perceived}')
```

```python
mean_creativity = df['Impact_on_Creativity'].mean()
print(f'Média da coluna Impact_on_Creativity: {mean_creativity}')
mode_creativity = df['Impact_on_Creativity'].mode()
print(f'Moda da coluna Impact_on_Creativity: {mode_creativity}')
std_creativity = df['Perceived_Effectiveness_of_VR'].std()
print(f'Desvio padrão da coluna Perceived_Effectiveness_of_VR: {std_creativity}')
```
